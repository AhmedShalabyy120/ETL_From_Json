```python
import numpy as np
import pandas as pd
import json
import datetime
```

```python
Records = pd.read_json('usa.gov_click_data.json' , lines = True)
```

```python
#Converting the json Data to DataFrame
df=pd.DataFrame(Records)
df.head(20)
```

```python
df['web_browser'] = df['a'].apply(lambda x : x.split('/')[0]) #getting the web browser from the first column
```

```python
df['web_browser']
```

```python
#getting the operating system name
df['os'] = df['a'].apply(lambda x : x.split('(')[1].split(';')[0] if '(' in x else None) 
```

```python
df['os']
```

```python
#getting the website which the cst came from 

df['from_url'] = df['r'].str.replace('http://','').str.replace('/+.*','', regex=True)
```

```python
df['from_url']
```

```python
#url cst heading to it 
df['to_url'] = df['u'].str.replace('https?://','', regex=True).str.replace('/+.*','', regex=True)
```

```python
df['to_url']
```

```python
df['city']=df['cy'].copy() #taking copy incase we changed anything in cy don't affect city 
```

```python
df.head()
```

```python
df['Latitude'] = df['ll'].str.get(0)
df['Latitude']
```

```python
df['longitude'] = df['ll'].str.get(1)
df['longitude']
```

```python
df['TimeZone']=df['tz'].copy()
```

```python
#using flash function to convert the time in
df['time in'] = pd.to_datetime(df['t'], unit='s').apply(lambda x: x.strftime('%Y-%m-%d %H:%M:%S'))
```

```python
#using a different way to get the time out
def convert_timestamp(ts):
    dt_object = datetime.datetime.fromtimestamp(ts)
    return dt_object.strftime('%Y-%m-%d %H:%M:%S')

df['time_out'] = df['hc'].apply(convert_timestamp)
df['time_out']
```

```python
#creating a new df with the columns we have mad
df_final=df.iloc[:,16:27]
```
